# Data Preprocessing

```python
# Checking null values
df.isnull().sum()
```

```python
import ast
```

```python
#Function to fetch genres and keywords
def convert(data):
  list = []
  for i in ast.literal_eval(data):
    list.append(i['name'])
  return list
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df['Keywords'] = df['Keywords'].apply(convert)
```

```python
# Working on cast column

def convertcast(data):
  list = []
  counter = 0
  for i in ast.literal_eval(data):
    if counter != 3:
      list.append(i['name'])
      counter+=1
    else:
      break
  return list
```

```python
df['cast'] = df['cast'].apply(convertcast)
```

```python
# Making function for director

def fetch_crew(data):
  list = []
  for i in ast.literal_eval(data):
    if i['job'] == 'Director':
      list.append(i['name'])
      break
  return list
```

```python
df['crew'] = df['crew'].apply(fetch_crew)
```

```python
df.head()
```

```python
# Converting overview from string to list
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
# Removing space from all words

df['Keywords'] = df['Keywords'].apply(lambda x:[i.replace(' ','') for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(' ','') for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(' ','') for i in x])
df['genres'] = df['genres'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
df.head()
```

```python
# Creating new column with all important columns concatenated

df['tag'] = df['overview'] + df['genres'] + df['Keywords'] + df['cast'] + df['crew']
```

```python
new_df = df[['imdb_id', 'title', 'tag']]
```

```python
new_df.head()
```

```python
# Converting tag column from list to string

new_df['tag'] = new_df['tag'].apply(lambda x:' '.join(x))
```

```python
#Converting to lower case
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

#Using Stemming and Count Vectorizer

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(string):
  list = []
  for i in string.split():
    list.append(ps.stem(i))
  return ' '.join(list)
```

```python
# Applying stemming on tag column
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.fit_transform(new_df['tag']).toarray().shape
```

```python
cv.get_feature_names()
```

```python
vectors
```

```python
# Using Cosine distance to find similarity between vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
similar[2]
```

```python
new_df[new_df['title'] == 'Kahaani'].index[0]
```

```python
sorted(list(enumerate(similar[5])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
# Function for recommending 5 similar movies
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distance = similar[movie_index]
  movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Transformers: The Movie')
```
